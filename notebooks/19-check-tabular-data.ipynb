```python
import pandas as pd
import glob, os
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_dir = '/glade/derecho/scratch/joko/synth-ros/params_200_50_20250403/tabular-data-v2'
# Use glob to find all .txt files in the directory
file_paths = glob.glob(f"{data_dir}/*.txt")

# Read all the files into a list of DataFrames and concatenate them
df = pd.concat([pd.read_csv(file, header=None) for file in file_paths], ignore_index=True)

# Display the final DataFrame
df.head()
```

```python
df.shape
```

```python
colnames = ['filename', 'unique_id', 'ros_id', 'proj_id', 'view', 'a', 'c', 
            'f_r0', 'f_hp', 'f_h0', 'n_arms', 'sa', 'vol', 
            'sa_eff', 'rho_eff', 'aspect_ratio', 'aspect_ratio_elip', 
            'extreme_pts', 'contour_area', 'contour_perimeter', 
            'area_ratio', 'complexity', 'circularity']
df.columns = colnames
```

```python
df.head()
```

```python
df.columns
```

```python
# Save to CSV with header
csv_filename = 'ros-tabular-data.csv'
csv_save_path = os.path.join(data_dir, csv_filename)
df.to_csv(csv_save_path, header=True, index=False)
```

```python
# test read CSV
df_test = pd.read_csv(csv_save_path)
df_test.head()
```

```python
# Save as parquet with header
pqt_filename = 'ros-tabular-data.parquet'
pqt_save_path = os.path.join(data_dir, pqt_filename)
df.to_parquet(pqt_save_path, index=False)
```

```python
# test read parquet
df_pqt = pd.read_parquet(pqt_save_path)
df_pqt.head()
```
